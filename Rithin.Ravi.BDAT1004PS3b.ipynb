Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\ravir\Downloads\us-marriages-divorces-1867-2014.csv')
```

Question 9

Create a pie chart showing the fraction of all Roman Emperors that were assassinated. Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.

```python
import pandas as pd

data = pd.read_csv(r'C:\Users\ravir\Downloads\roman-emperor-reigns.csv')

data.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

new_data = data.groupby("Cause_of_Death")['Emperor'].count()
new_data
```

```python
new_data.plot.pie(autopct="%.1f%%", figsize = (10,10) , fontsize = 22, ylabel = '', explode=[0.01]*8)
plt.title("Roman Emperor's Cause of Death", color = 'Black', fontsize = 20, fontweight = 'bold')
```

Question 10

Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009.

```python
import pandas as pd

data = pd.read_csv(r'C:\Users\ravir\Downloads\arcade-revenue-vs-cs-doctorates.csv')

data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

colours = ['#C4E57D','#2AC280','#FFAE39','#DC5B3B','#52E397','#C5CBA3','#9CD5F6','#6E50D9','#9A5E59','#9BC8F5']

data.plot.scatter(x='Total Arcade Revenue (billions)',
                  y='Computer Science Doctorates Awarded (US)', c=colours , s = 50, figsize = (5,4))

plt.title('Revenue Vs CS Doctorates', color = 'Black', fontsize = 16)
plt.xlabel('Total Arcade Revenue (billions)' , color = 'Black', fontsize = 12)
plt.ylabel('CS Doctorates Awarded', color = 'Black', fontsize = 10)
```
